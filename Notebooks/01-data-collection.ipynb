```python
import requests
import zipfile
import io
import os

# --- Configuration ---
DATASET_URL = 'https://archive.ics.uci.edu/static/public/319/mhealth+dataset.zip'
RAW_DATA_DIR = 'data/raw'

# Ensure directory exists
os.makedirs(RAW_DATA_DIR, exist_ok=True)

def download_and_extract():
    print(f"Downloading dataset from {DATASET_URL}...")
    try:
        response = requests.get(DATASET_URL)
        response.raise_for_status()

        print("Extracting files...")
        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            z.extractall(RAW_DATA_DIR)

        print(f"Success! Data extracted to '{RAW_DATA_DIR}'")

        # Verify extraction
        files = os.listdir(os.path.join(RAW_DATA_DIR, 'MHEALTHDATASET'))
        print(f"Found {len(files)} files in dataset folder.")

    except Exception as e:
        print(f"Error during download/extraction: {e}")

if __name__ == "__main__":
    download_and_extract()
```
